# 5. Multivariate Analysis

```python
# Total Spending by Category and Store Type
# The sources indicate a mix of In-store and Online transactions across various categories [1, 7, 8].
category_pivot = df.pivot_table(
    values='Total Spent', 
    index='Category', 
    columns='Location', 
    aggfunc='sum'
)
print("\nTotal Spending by Category and Channel:\n", category_pivot)
```

```python
# Visualizing Discount Applied Impact on Purchase Quantities
# Data snippets show that Discount Applied often impact the quantities purchased [1, 2, 9].
# Encode Discount Applied to numeric for correlation
if "Discount Applied" in df.columns:
    disc_map = {True: 1, False: 0, "True": 1, "False": 0, "Yes": 1, "No": 0, "Unknown": np.nan}
    df["Discount Applied Num"] = df["Discount Applied"].map(disc_map)

numeric_df = df.select_dtypes(include=["number"]).copy()

corr_mat = numeric_df.corr(numeric_only=True)
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Quantity', hue='Discount Applied Num', data=df)
plt.title('Impact of Discount Applied Num on Quantity across Categories')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

January stands out with the highest activity (~1300), while other months remain steady (~950–1100).
