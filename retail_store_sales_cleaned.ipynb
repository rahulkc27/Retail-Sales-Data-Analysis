###  Handle Missing Values

```python
df.head()
```

```python
df['Item'] = df['Item'].fillna("Unknown")
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Total Spent'] = df['Total Spent'].fillna(df['Total Spent'].median())
df['Discount Applied'] = df['Discount Applied'].fillna("Unknown")
```

```python
df.isna().sum()
```

```python
df
```

### Remove Duplicates

```python
df = df.drop_duplicates()
```

### Correct Data Types

```python
df.head()
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
df['Quantity'] = df['Quantity'].astype(int)
```

```python
df.dtypes
```

### Create Derived Columns

```python
df['Revenue'] = df['Price Per Unit'] * df['Quantity']
```

```python
df['Month'] = df['Transaction Date'].dt.month
df['Year'] = df['Transaction Date'].dt.year
```

### Filter or Aggregate Data

```python
df = df[df['Price Per Unit'] > 0]   # remove invalid sales
```

```python
monthly_sales = df.groupby(['Year','Month'])['Revenue'].sum().reset_index()
```
